```python
import glob
import json
import pandas as pd
```

```python
ds = pd.read_csv('data/baseline/enriched_dataset.csv', index_col = 0)
```

```python
report_ds = {}

paths = glob.glob('data/reports/*')
for path in paths:
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            report = json.loads(line)
            if (report["name"], int(report["date"]) - 1) not in report_ds:
                report_ds[(report["name"], int(report["date"]) - 1)] = []
            report_ds[(report["name"], int(report["date"]) - 1)].append(report["content"])
tmp = {
    "ticker": [],
    "year": [],
    "reports": []
}
for k in report_ds:
    tmp["ticker"].append(k[0])
    tmp["year"].append(k[1])
    tmp["reports"].append(report_ds[k])
report_ds = pd.DataFrame(tmp)
```

```python
ds = pd.merge(ds, report_ds, on=["ticker", "year"], suffixes=(None, None), how="left").to_csv("/Users/vencel/diploma/data/enriched_dataset.csv")
```
